```python
bronze_table = f"{catalog_name}.{bronze_schema}.constructors"
silver_table = f"{catalog_name}.{silver_schema}.constructors"
```

```python
read_constructors_df = spark.table(bronze_table)
```

```python
from pyspark.sql import functions as F
constructors_select_df = read_constructors_df.select(
F.col("constructorId").alias("constructor_id"),
F.col("name").alias("constructor_name"),
F.col("nationality"),
F.col("ingestion_timestamp"),
F.col("source_file")

)
```

```python
constructors_rm_duplicate = constructors_select_df.dropDuplicates(["constructor_id"])
```

```python
from pyspark.sql import functions as F

constructors_final_df= (constructors_rm_duplicate
                      .withColumn(("nationality"),F.initcap(F.col("nationality")))
)

```

```python
(constructors_final_df
 .write
 .format("delta")
 .mode("overwrite")
 .saveAsTable(silver_table)
)
```
