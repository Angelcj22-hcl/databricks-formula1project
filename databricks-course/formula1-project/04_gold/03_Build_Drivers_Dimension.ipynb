```python
drivers_df               = spark.table(f"{catalog_name}.{silver_schema}.drivers")
ref_nationality_region_df = spark.table(f"{catalog_name}.{gold_schema}.ref_nationality_region")
```

```python
target_table = f"{catalog_name}.{gold_schema}.dim_drivers"
```

```python
constructors_df = spark.table(f"{catalog_name}.{silver_schema}.drivers")
ref_nationality_df = spark.table(f"{catalog_name}.{gold_schema}.ref_nationality_region")  
```

```python
dim_drivers_df = (
                    drivers_df
                    .join(
                        ref_nationality_df, 
                        drivers_df.nationality== ref_nationality_df.nationality,
                        "left_outer"
                          )
                    .select(
                        drivers_df.driver_id,
                        drivers_df.driver_name,
                        drivers_df.date_of_birth,
                        drivers_df.nationality,
                        ref_nationality_df.region.alias("nationality_region"),
                        
                    )
)
```

```python
(dim_drivers_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
    )
```
